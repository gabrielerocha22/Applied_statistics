```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
dados = pd.DataFrame({'Profissão': [1, 2, 3, 1, 2, 2, 2, 3, 3, 2, 1, 3]})
```

```python
# Frequências
frequencia = dados['Profissão'].value_counts()

# Ordenando as categorias
frequencia = frequencia.sort_index()

# Criando uma tabela com as frequências e os percentuais
tabela = pd.DataFrame({'Frequência': frequencia, 'Percentual': frequencia / len(dados) * 100})

# Adicionando o nome das categorias
tabela.index.name = 'Profissão'
tabela.rename(index={1: 'Estatístico', 2: 'Cientista de Dados', 3: 'Programador Python'}, inplace=True)

# Tabela
print(tabela)
```

```python
# Gráfico - Tamanho 
plt.figure(figsize=(6, 4))

# Histograma
plt.hist(dados['Profissão'], bins=[1, 2, 3, 4], align='left', rwidth=0.5, color='steelblue')
plt.xlabel('Profissão')
plt.ylabel('Frequência')

# Categorias:
plt.xticks([1, 2, 3], ['Estatístico', 'Cientista de Dados', 'Programador Python'])

# Gráfico
plt.show()
```
